```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies=pd.read_csv('imdb_movies.csv')
movies
```

```python
movies.info()
```

```python
movies.isnull().sum(axis=0).sort_values(ascending=False)
```

```python
movies.isnull().sum(axis=1).sort_values(ascending=False)
```

Delete unwanted cells

```python
movies=movies.drop(['crew'],axis=1)
```

```python
movies.isnull().sum().sort_values(ascending=False)/len(movies)
```

Remove the rows which contain null values

```python
movies=movies[movies['genre'].notnull()]
movies
```

```python
movies.isnull().sum().sort_values(ascending=False)/len(movies)
```

```python
(movies.isnull().sum(axis=1).sort_values(ascending=False)>5).sum()
#no. of rows which have less than 5 values
```

```python
movies=movies[movies.isnull().sum(axis=1).sort_values(ascending=False)<=5]
movies
```

Profit

```python
#movies with Highest profit
movies['Profit']=movies['revenue']-movies['budget']
movies
```

```python
#converting units from $ to million$
movies['budget']=movies['budget']/1000000
movies['revenue']=movies['revenue']/1000000
movies
```

```python
movies.sort_values(by='Profit',ascending=False)
```

Top 10 Movies

```python
#remove duplicate values
movies.drop_duplicates(keep='first',inplace=True)
top10=movies.sort_values(by='Profit',ascending=False).head(10)
top10
```

```python
movies=movies.sort_values(by='score',ascending=False)
movies
```

```python
imdb_top_250=movies.sort_values(by='score',ascending=False).head(250)
imdb_top_250
```

Rank acc to IMDB score

```python
movies['Rank']=movies['score'].rank(method='first',ascending=False)
movies
```

```python
#find popular genre
tempGenre=movies.genre.str.split(',',expand=True).iloc[:,0:2]
tempGenre.columns=['genre1','genre2']
tempGenre
```

```python
tempGenre.genre2.fillna(tempGenre.genre1,inplace=True)
tempGenre
```

```python
movies=pd.concat([movies,tempGenre],axis=1)
movies
```

TOP 10 genre

```python
#top 10 Genre
movies.groupby(['genre1','genre2']).revenue.mean().sort_values(ascending=False).head(10)
```
